```python
import pandas as pd

df = pd.read_csv('data.csv')
df.describe()
df.head()
```

```python
import matplotlib.pyplot as plt

y = list(df.groupby('cyan')['time'])
plt.boxplot(list(map(lambda x: x[1], y)), tick_labels=list(map(lambda x: x[0], y)))
plt.xlabel('cyan')
plt.ylabel('time')
plt.show()
```
